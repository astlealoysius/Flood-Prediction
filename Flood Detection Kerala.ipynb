```python
import os
data_path = ['data']
```

```python
import pandas as pd

# Import the data using the file path
filepath = os.sep.join(data_path + ['kerala.csv'])
data = pd.read_csv('kerala.csv')
data
```

```python
data.drop('SUBDIVISION', axis=1, inplace=True)
data.fillna(0, inplace=True)
print(data)
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()

for col in ['FLOODS']:
    data[col] = lb.fit_transform(data[col])
```

```python
data
```

```python
x = data.loc[:, 'YEAR':'DEC']  
y = data.loc[:, 'FLOODS']
# Display the extracted columns
print(x)
```

```python
print(y)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = data[['JAN', 'FEB', 'MAR', 'APR','MAY', 'JUN', 'AUG', 'SEP', 'OCT','NOV','DEC']].mean().plot.bar(
    width=0.6, edgecolor='black', align='center', linewidth=2, figsize=(12, 6), color='skyblue')

# Labeling the plot and setting axis labels
plt.xlabel('Months', fontsize=15)
plt.ylabel('Mean Monthly Rainfall', fontsize=15)
plt.title('Average Monthly Rainfall in Kerala', fontsize=18)
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability

# Adding data labels to the bars
for i in ax.patches:
    ax.text(i.get_x() + i.get_width() / 2, i.get_height() + 0.1, f'{i.get_height():.2f}', ha='center', va='bottom')

# Additional styling
plt.grid(axis='y')
sns.set(style='whitegrid')  # Using a different style for the plot

plt.show()
```

```python
# Importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearnex import patch_sklearn
import warnings

# Enable Intel Extension for Scikit-learn
patch_sklearn()

# Initialize the Min-Max scaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(x)

# Split the scaled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train the logistic regression model
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(X_train, y_train)

# Make predictions
y_train_pred_binary = logistic_regression_model.predict(X_train)
y_test_pred_binary = logistic_regression_model.predict(X_test)

# Calculate accuracy
train_accuracy = accuracy_score(y_train, y_train_pred_binary)
test_accuracy = accuracy_score(y_test, y_test_pred_binary)

print(f'Training Set Accuracy: {train_accuracy * 100:.2f}%')
print(f'Testing Set Accuracy: {test_accuracy * 100:.2f}%')
```

```python
# Calculate precision, recall, and F1 score
precision = precision_score(y_test, y_test_pred_binary)
recall = recall_score(y_test, y_test_pred_binary)
f1 = f1_score(y_test, y_test_pred_binary)

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```

```python
# Calculate confusion matrix for testing set
cm = confusion_matrix(y_test, y_test_pred_binary)

# Create a heatmap of the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
